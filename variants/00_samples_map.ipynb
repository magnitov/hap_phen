```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
plt.figure(figsize = (9, 5))
m = Basemap(projection = 'robin',lon_0 = 0, resolution = 'c')
m.drawcoastlines()
m.fillcontinents(color = 'whitesmoke', lake_color = 'white')
m.drawparallels(np.arange(-90, 120, 30), linewidth = 0.5)
m.drawmeridians(np.arange(0, 360, 60), linewidth = 0.5)
m.drawmapboundary(fill_color = 'white')

# Annotate locations
ns = [39.3210, 36.2048, 18.2208, 30.3753, 8.4606]
ew = [-111.0937, 138.2529, -66.5901, 69.3451, -11.7799]

x, y = m(ew, ns)
m.scatter(x, y, color = ['C0', 'C2', 'C3', 'C4', 'C1'], s = 150, edgecolor = 'k', zorder = 3)

# Annotate sample names
x, y = m([-151, 127, -73, 69, 7], [30, 26, 28, 39, -1])

plt.text(x[0], y[0], 'NA12878', fontsize = 16, fontweight = 'bold', ha = 'left', va = 'center', 
         color = 'k', bbox = {'boxstyle': 'square,pad=0.1', 'color': 'C0', 'alpha': 0.5})
plt.text(x[1], y[1], 'NA18983', fontsize = 16, fontweight = 'bold', ha = 'left', va = 'center', 
         color = 'k', bbox = {'boxstyle': 'square,pad=0.1', 'color': 'C2', 'alpha': 0.5})
plt.text(x[2], y[2], 'HG01241', fontsize = 16, fontweight = 'bold', ha = 'left', va = 'center', 
         color = 'k', bbox = {'boxstyle': 'square,pad=0.1', 'color': 'C3', 'alpha': 0.5})
plt.text(x[3], y[3], 'HG02601', fontsize = 16, fontweight = 'bold', ha = 'left', va = 'center', 
         color = 'k', bbox = {'boxstyle': 'square,pad=0.1', 'color': 'C4', 'alpha': 0.5})
plt.text(x[4], y[4], 'HG03464', fontsize = 16, fontweight = 'bold', ha = 'right', va = 'center', 
         color = 'k', bbox = {'boxstyle': 'square,pad=0.1', 'color': 'C1', 'alpha': 0.5})


plt.tight_layout()
plt.savefig('/DATA/users/m.magnitov/hap_phen/figures/samples_globe.pdf', bbox_inches = 'tight')
plt.show()
```
